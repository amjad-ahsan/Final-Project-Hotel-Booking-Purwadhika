##### **F. *Confusion Matrix***

*Confusion Matrix* adalah tabel yang digunakan untuk mengevaluasi kinerja model klasifikasi dengan membandingkan hasil prediksi dengan label sebenarnya. Dalam proyek ini, *Confusion Matrix* digunakan sebagai acuan dalam pengambilan keputusan pada penerapan Machine Learning.

---

|                    | Prediksi Positif   | Prediksi Negatif   |
| ------------------ | ------------------ | ------------------ |
| **Aktual Positif** | TP (benar positif) | FN (salah negatif) |
| **Aktual Negatif** | FP (salah positif) | TN (benar negatif) |

---

**Keterangan:**

- **TP** (True Positive): Pemesanan yang dibatalkan dan berhasil diprediksi akan dibatalkan oleh model.

- **TN** (True Negative): Pemesanan yang tidak dibatalkan dan berhasil diprediksi tidak dibatalkan oleh model.

- **FP** (False Positive): Pemesanan yang sebenarnya tidak dibatalkan, tetapi model memprediksinya akan dibatalkan.

- **FN** (False Negative): Pemesanan yang sebenarnya dibatalkan, tetapi model memprediksinya tidak dibatalkan.

**Metrik dari Confusion Matrix**:

* Akurasi = (TP + TN) / Total
* Presisi = TP / (TP + FP)
* Recall = TP / (TP + FN)
* F1-Score = 2 × (Presisi × Recall) / (Presisi + Recall)

**Kesalahan yang Paling Merugikan**

Dalam proyek ini, Matriks Confsuon yang paling merugikan adalah **FP** atau False Positive. Karena jika sebuah pemesanan yang sebenarnya tidak batal diprediksi akan dibatalkan oleh pihak Hotel, manajemen hotel mungkin mengambil tindakan pencegahan seperti pembatalan reservarsi atau mengalihkan tamu hotel ke hotel lain.

Hal ini dapat menyebabkan:
- Kehilangan pendapatan karena kamar sudah disiapkan namun tidak digunakan.
- Menurunya reputasi hotel karena tidak dapat memenuhi pemesana pelanggan.
- Potensi hotel mengalami overbooking padahal hotel sebenarnya tidak.

**Kenapa Bukan False Negative?**

Karena jika terjadi False Negative, masih ada kemungkinan kamar tersebut dapat segera dipasarkan ulang — terutama jika dibatalkan mendadak.

Meskipun tetap berpotensi mengurangi pendapatan, dampaknya cenderung lebih kecil dibanding kerugian akibat FP, karena hotel masih memiliki kesempatan mitigasi lebih cepat dengan langkah berikut:
- Dengan menghubungi tamu secara langsung untuk memastikan kehadiran mereka sebelum periode reservasi berakhir jika tidak berkunjung dalam waktu reservarsi hotel.
- Mempromosikan kamar secara cepat melalui online travel agent (OTA) atau website hotel untuk menarik pemesan baru.
- Menggunakan daftar tunggu (waiting list) dari calon tamu yang sebelumnya tidak mendapatkan kamar.
- Memberikan diskon menit terakhir (last-minute deals) guna meningkatkan peluang kamar segera terjual kembali.

---
| Jenis Error        | Sumber Nominal Kerugian     | Cara Menghitung                                                       |  Jumlah kerugian                           |
|--------------------|-----------------------------|----------------------------------------------------------------------|--------------------------------------------|
| False Negative (FN)| Pendapatan yang Hilang      | Jumlahkan (adr * total_malam_menginap) untuk setiap kasus FN.      |  €3.43/malam - deposit                        |
| False Positive (FP)| Biaya Operasional           | Jumlahkan (biaya_per_tindakan) untuk setiap kasus FP.               | €3.43/malam (karena kehilangan 1 room)              |
---


Dengan demikian, proyek ini akan berfokus pada prediksi False Positive (FP) guna meminimalkan potensi kerugian yang dialami pihak hotel.

#### **A. Menambah Feature Tambahan**

```python


# 1. Total nights stayed
df['total_nights'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']

# 2. Total guests
df['total_guest'] = df['adults'] + df['children'] + df['babies']

# 3. Has children (hanya cek children atau babies, TANPA syarat dewasa)
df['has_children'] = np.where((df['children'] > 0) | (df['babies'] > 0), 1, 0)

# 4. Is peak season (June-August)
peak_months = ['June', 'July', 'August']
df['is_peak_season'] = df['arrival_date_month'].isin(peak_months).astype(int)

# 5. Lead time categories
conditions = [
    (df['lead_time'] <= 7),
    (df['lead_time'] > 7) & (df['lead_time'] <= 30),
    (df['lead_time'] > 30) & (df['lead_time'] <= 90),
    (df['lead_time'] > 90)
]
choices = ['last_minute', 'short_term', 'medium_term', 'long_term']
df['lead_time_category'] = np.select(conditions, choices, default='unknown')

# 6. Fixed refund (deposit non-refundable)
df['high_risk_deposit'] = np.where(df['deposit_type'].isin(['Non Refund']), 1, 0)

# 7. Is domestic (PRT)
df['is_domestic'] = np.where(df['country'] == 'PRT', 1, 0)

df['cancel_ratio'] = df['previous_cancellations'] / (df['previous_bookings_not_canceled'] + 1)

df['last_minute_booking'] = (df['lead_time'] <= 3).astype(int)


# Simpan ke file CSV jika diperlukan
# df.to_csv('hotel_bookings_with_features.csv', index=False)
```

#### **C. Outliers**

Langkah selanjutnya adalah melakukan pemeriksaan terhadap data untuk mengidentifikasi keberadaan outlier. Outlier adalah nilai yang menyimpang jauh dari distribusi normal data dan berpotensi memengaruhi kinerja model. Setelah teridentifikasi, outlier yang tidak relevan atau disebabkan oleh kesalahan pencatatan akan dihapus agar data yang digunakan lebih bersih, representatif, dan dapat meningkatkan akurasi model.



```python
import math

numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns

n_cols = 4
n_rows = math.ceil(len(numerical_columns) / n_cols)

plt.figure(figsize=(5 * n_cols, 4 * n_rows))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(n_rows, n_cols, i)
    sns.boxplot(x=df[col], color='skyblue')
    plt.title(f'{col} Outliers', fontsize=12)
    plt.tight_layout()
plt.show()
```

Berdasarkan hasil analisis distribusi, ditemukan bahwa fitur `lead_time`, `booking_changes`, `previous_bookings_not_canceled`, `previous_cancellations`, `stays_in_week_nights`, `stays_in_weekend_nights`, `adults`, dan `total_guest_count` memiliki nilai yang tergolong *outlier*. Nilai-nilai ini berada jauh di luar rentang normal mayoritas data, sehingga berpotensi memengaruhi hasil analisis dan kinerja model jika tidak ditangani dengan tepat.


```python
cols = [
    'lead_time',
    'booking_changes',
    'previous_bookings_not_canceled',
    'previous_cancellations',
    'stays_in_week_nights',
    'stays_in_weekend_nights',
]
for col in cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outlier_count = ((df[col] < lower_bound) | (df[col] > upper_bound)).sum()

    print(f"{col}")
    print(f"  Q1: {Q1:.2f}")
    print(f"  Q3: {Q3:.2f}")
    print(f"  IQR: {IQR:.2f}")
    print(f"  Lower Bound: {lower_bound:.2f}")
    print(f"  Upper Bound: {upper_bound:.2f}")
    print(f"  Outliers Detected: {outlier_count} ({outlier_count / len(df):.2%})")
    print("-" * 50)
```

Setelah dilakukan pembahasan, diputuskan untuk memberikan *capping* pada fitur `lead_time`, `stays_in_weekend_nights`, dan `adr`. *Capping* dilakukan untuk membatasi nilai ekstrem (*extreme values*) pada batas maksimum tertentu sehingga distribusi data menjadi lebih stabil, mengurangi pengaruh *outlier* yang terlalu besar, serta menjaga informasi penting tanpa harus menghapus data secara keseluruhan.


```python
# hapus outlier dari adr dengan batas di 1000
df = df[df['adr'] <= 1000]

# hapus outlier dari lead_time
df = df[df['lead_time'] <= 296]

# hapus outlier dari stays_in_weekend_nights dengan batas di 100
df = df[df['stays_in_weekend_nights'] <= 5]
```

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_results = []  # CV metrics
test_results = []  # Final test metric
```

```python
def evaluate_model(pipeline, X, y):
    cv_metrics = cross_validate(
        pipeline, X, y, cv=skf,
        scoring=['f1', 'precision', 'average_precision'],  # Changed metrics
        return_train_score=True,
        n_jobs=-1
    )
    return {
        'F0.5 (CV)': np.mean(cv_metrics['test_f1']),
        'Precision (CV)': np.mean(cv_metrics['test_precision']),
        'PRC AUC (CV)': np.mean(cv_metrics['test_average_precision'])  # PRC AUC
    }

results = []
for name, model in models.items():
    pipeline = ImbPipeline([
        ('preprocessing', make_preprocessor()),
        ('smote', SMOTE(random_state=42)),
        ('classifier', model)
    ])
    
    # CV evaluation
    cv_metrics = evaluate_model(pipeline, X_train, y_train)
    
    # Test evaluation
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)
    y_proba = pipeline.predict_proba(X_test)[:, 1]
    
    # Calculate specificity
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
    specificity = tn / (tn + fp)
    
    results.append({
        'Model': name,
        **cv_metrics,
        'F0.5 (Test)': fbeta_score(y_test, y_pred, beta=0.5),
        'Precision (Test)': precision_score(y_test, y_pred, pos_label=1),
        'Specificity (Test)': specificity,
        'PRC AUC (Test)': average_precision_score(y_test, y_proba)  # PRC AUC
    })

# Create dataframe and reorder columns
results_df = pd.DataFrame(results).set_index('Model')
results_df = results_df[[
    'Precision (CV)', 'F0.5 (CV)', 'PRC AUC (CV)',
    'Precision (Test)', 'F0.5 (Test)', 'Specificity (Test)', 'PRC AUC (Test)'
]]
```

### **ENSEMBLE XGB dan RF (FP FOCUS)**

```python
print("ENSEMBLE MODEL (FP Focus) - Tuned RF + XGB")
# Ensemble Model
ensemble = VotingClassifier(
    estimators=[
        ('rf', best_rf),
        ('xgb', xgb_untuned)
    ],
    voting='soft',
    n_jobs=-1
)

# Train
ensemble.fit(X_train, y_train)

# Probabilities
y_proba_ensemble = ensemble.predict_proba(X_test)[:, 1]

# FP-focused threshold
threshold_fp_focus = 0.7

def evaluate_model(y_true, y_pred, y_proba, label):
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f05 = fbeta_score(y_true, y_pred, beta=0.5)  # Fokus precision
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    specificity = tn / (tn + fp)
    fp_rate = fp / (fp + tn)
    prc_auc = average_precision_score(y_true, y_proba)

    print(f"\n--- {label} ---")
    print(classification_report(y_true, y_pred))
    print(f"Precision      : {precision:.4f}")
    print(f"Recall         : {recall:.4f}")
    print(f"F0.5 Score     : {f05:.4f}")
    print(f"Specificity    : {specificity:.4f}")
    print(f"False Positives: {fp} (FP Rate: {fp_rate:.4%})")
    print(f"PRC AUC        : {prc_auc:.4f}")
    return fp

# FP-focused threshold only
y_pred_fp_focus = (y_proba_ensemble >= threshold_fp_focus).astype(int)
fp_count = evaluate_model(y_test, y_pred_fp_focus, y_proba_ensemble, f"FP-Optimized Threshold {threshold_fp_focus}")

# Visualisasi CM untuk threshold 0.7
ConfusionMatrixDisplay.from_predictions(
    y_test, y_pred_fp_focus,
    cmap="Greens",
    display_labels=['Negative', 'Positive'],
    values_format='d'
)
plt.title(f"Confusion Matrix - Ensemble (Threshold {threshold_fp_focus})")
plt.show()

# PRC Curve
PrecisionRecallDisplay.from_predictions(
    y_test, y_proba_ensemble,
    color='green'
)
plt.title("Precision-Recall Curve - Ensemble (Soft Voting)")
plt.show()
```

**Fokus Utama:**  
- **Meminimalkan False Positive (FP)** (prediksi "batal" yang salah), karena kesalahan ini paling merugikan hotel (misalnya kamar kosong akibat tindakan pencegahan yang tidak perlu).

3. **Optimalisasi Pendapatan:** Meningkatkan hunian kamar dan mengurangi kerugian akibat pembatalan mendadak.

- **FP (False Positive)** lebih berbahaya daripada FN (False Negative) karena menyebabkan kamar tidak terjual padahal tamu sebenarnya tidak membatalkan.  
- **FN (False Negative)** masih bisa dimitigasi dengan langkah last-minute seperti penjualan ulang kamar atau daftar tunggu.
